# 주제 : Black Sabbath IRONMAN on Tchibo Toy Guitar



```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
#1. 브라우저를 실행 > 유튜브채널 사이트로!
driver = wb.Chrome(f'C:/chrome/chromedriver.exe')   
b="https://www.youtube.com/watch?v=CZzad1e1D9w"
driver.get(b)
```

```python
#2. 댓글정보를 수집
review = driver.find_elements(By.ID,"content-text")
all = ""
for i in review :
    print(i.text)
    all += i.text
    
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))


word_tokens = word_tokenize(str(all))  
filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]

filtered_sentence = []
for w in word_tokens:
    if w not in stop_words:
        new_string = ''.join(char for char in w if char.isalnum())
        if new_string != '' and new_string != 'I' and 1 != len(new_string) :
            filtered_sentence.append(new_string)
        
print(filtered_sentence)
```

```python
from wordcloud import WordCloud # 클래스
wc = WordCloud(background_color="white", 
               font_path="C:\Windows\Fonts\malgunsl.ttf",
               random_state=802)
reviews = " ".join(filtered_sentence) # 하나의 문자열로 변경
cloud = wc.generate_from_text(reviews)
import matplotlib.pyplot as plt
plt.figure(figsize=(20,8))
plt.imshow(cloud)
plt.axis("off") # x축,y축 끄기
plt.savefig("./보고서9.jpg") # 그림저장
plt.show()
```

# 단어가 있는것들중에 상위 15개 단어 시각화 하기

```python
from collections import Counter
wordcount = Counter(filtered_sentence)
print(wordcount.most_common())
temp = wordcount.copy() #명사들만 호출한거 temp에 넣기
for i in temp:         # temp 안에 한개씩 호출
    if len(i) == 1:    # 길이가 한개인 단어일경우
        del wordcount[i]#제거함
import collections
import matplotlib.pyplot as plt
count = dict(Counter(wordcount).most_common(15)) #상위 15개 뽑기
print(count)
plt.figure(figsize=(15,15), dpi=80)
plt.bar(count.keys(), count.values())
plt.savefig("./보고서9plt.jpg")
```
