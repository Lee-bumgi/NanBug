```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
#1. 브라우저를 실행 > 유튜브채널 사이트로!
driver = wb.Chrome(f'C:/chrome/chromedriver.exe')   
a= "https://www.youtube.com/watch?v=zxKUGIE6bM0"
driver.get(a)
```

```python
#2. 댓글정보를 수집
review = driver.find_elements(By.ID,"content-text")
all = ""
for i in review :
    print(i.text)
    all += i.text
    
```

# 단어가 있는것들중에 상위 15개 단어 출력하기

```python
from konlpy.tag import Okt
okt = Okt() #코넬피 호출
okt.nouns  #코넬피.명사
nounsword = okt.nouns(all) #전체 댓글중에 명사들만 호출하기 
temp = nounsword.copy() #명사들만 호출한거 temp에 넣기
for i in temp:         # temp 안에 한개씩 호출
    if len(i) == 1:    # 길이가 한개인 단어일경우
        nounsword.remove(i) #제거함
from collections import Counter
count = dict(Counter(nounsword).most_common(15)) #명사 갯수는 몇개?
import matplotlib.pyplot as plt #매트플로리프 호출
from matplotlib import font_manager, rc #한글 되는거 호출
plt.rcParams['axes.unicode_minus'] = False


font_path = "c:/Windows/Fonts/malgun.ttf" #한글 폰트 부르기

font_name = font_manager.FontProperties(fname=font_path).get_name() #한글 폰트 불러서 적용준비하기
rc('font', family=font_name) #적용하기

count = dict(Counter(nounsword).most_common(15)) #상위 15개 뽑기
print(count)
plt.figure(figsize=(15,15), dpi=80)
plt.bar(count.keys(), count.values())
plt.savefig("./보고서1plt.jpg") # 그림저장
```

```python
from wordcloud import WordCloud # 클래스
wc = WordCloud(background_color="white", 
               font_path="C:\Windows\Fonts\malgunsl.ttf",
               random_state=802)
reviews = " ".join(nounsword) # 하나의 문자열로 변경
cloud = wc.generate_from_text(reviews)
import matplotlib.pyplot as plt
plt.figure(figsize=(20,8))
plt.imshow(cloud)
plt.axis("off") # x축,y축 끄기
plt.savefig("./보고서1.jpg") # 그림저장
plt.show()
```
